### 데이터 통합 (변수 별)

```python
def make_dataset(year=2018):
    
    if year == 2019:
        data_selected = [data_2019_temperature_3H, data_2019_sky, 
                         data_2019_rainfall_probability, data_2019_rainfall_6H, data_2019_snow_6H, 
                         data_2019_humidity, data_2019_wind_speed, data_2019_wind_direction]
    elif year == 2018:
        data_selected = [data_2018_temperature_3H, data_2018_sky, 
                         data_2018_rainfall_probability, data_2018_rainfall_6H, data_2018_snow_6H, 
                         data_2018_humidity, data_2018_wind_speed, data_2018_wind_direction]
    elif year == 2017:
        data_selected = [data_2017_temperature_3H, data_2017_sky, 
                         data_2017_rainfall_probability, data_2017_rainfall_6H, data_2017_snow_6H, 
                         data_2017_humidity, data_2017_wind_speed, data_2017_wind_direction]
    elif year == 2016:
        data_selected = [data_2016_temperature_3H, data_2016_sky, 
                         data_2016_rainfall_probability, data_2016_rainfall_6H, data_2016_snow_6H, 
                         data_2016_humidity, data_2016_wind_speed, data_2016_wind_direction]
    else:
        print("No selected year.")
    
    data_concat = pd.concat(data_selected, axis=1)

    # make day_, hour_, forecast_ and drop ...
    data_concat['day_'] = data_concat.iloc[:, :1].values
    data_concat['hour_'] = data_concat.iloc[:, 1:2].values
    data_concat['forecast_'] = data_concat.iloc[:, 2:3].values
    data_concat.drop([' format: day', 'hour', 'forecast'], axis=1, inplace=True)

    # change columns
    data_concat.columns = ['temperature_3H', 'sky', 'rainfall_probability', 'rainfall', 'snow', 'humidity', 
                           'wind_speed', 'wind_direction', 'day', 'hour', 'forecast']

    # fillna
    data_concat[['rainfall', 'snow']] = data_concat[['rainfall', 'snow']].fillna(0)

    # drop Nan Value
    data_concat = data_concat.dropna()
    
    return data_concat
```

---

### 8AM 선택, 날짜변경 (예보시기 -> 실제 시기)

```python
year_name = [2016, 2017, 2018, 2019]

for year in year_name:

    # choose hour 8
    data_concat = make_dataset(year=year)
    data_concat = data_concat[data_concat.hour == 800.0]
    data_concat = data_concat[data_concat.forecast <= 25]
    print(data_concat.shape)

    # copy data_concat
    m = data_concat.shape[0]

    # reset index
    data_concat.reset_index(drop=True, inplace=True)

    # 실제 시기
    n = 4
    if year == 2016:
        data_concat.insert(loc=0, column='time', value=pd.date_range('{}-10-01'.format(str(year)), periods=m + n, freq='3H')[n:]) 
    else:
        data_concat.insert(loc=0, column='time', value=pd.date_range('{}-01-01'.format(str(year)), periods=m + n, freq='3H')[n:]) 

    # drop day, hour, forecast
    data_concat.drop(['hour', 'day', 'forecast'], axis=1, inplace=True)

    # save dataset(pickle)
    data_concat.to_pickle('/Users/quartz/GoogleDrive/ewp/forecast_dangjin/data_forecast_{}_AM_06.pkl'.format(year))
```
