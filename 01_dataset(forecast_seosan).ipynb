### 데이터 통합 (변수 별)

```python
data_concat = pd.concat([data_2016_temperature_3H, data_2016_sky, data_2016_precipitation_form, 
                         data_2016_rainfall_probability, data_2016_humidity, 
                         data_2016_wind_speed, data_2016_wind_direction], axis=1)
```

```python
# drop Nan Value
data_concat.dropna(inplace=True)

# change dtypes
data_concat[' format: day'] = data_concat[' format: day'].astype('float64')

# make day_, hour_, forecast_ and drop ...
data_concat['day_'] = data_concat.iloc[:, :1].values
data_concat['hour_'] = data_concat.iloc[:, 1:2].values
data_concat['forecast_'] = data_concat.iloc[:, 2:3].values
data_concat.drop([' format: day', 'hour', 'forecast'], axis=1, inplace=True)

# change columns
data_concat.columns = ['temperature_3H', 'sky', 'precipitation_form', 'rainfall_probability', 'humidity', 'wind_speed', 'wind_direction', 'day', 'hour', 'forecast']
data_concat.tail()
```

### 노트

- 2017년 예보 데이터의 하늘 상태(sky), 강수 형태(pre... form) 데이터가 부족하다.
- 2017년 하늘상태 데이터가 5월 ~ 12월 부분이 누락되어 있다.
- 2017년 강수형태 데이터는 거의 유사. 차후 확인 필요.
