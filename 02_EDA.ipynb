```python
# 다중공선성 확인

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["feature"] = X.columns
vif.sort_values(by="VIF Factor", ascending=False)
```

```python
model.pvalues[model.pvalues <= 0.01]
```

```python
# 자동으로 top vif factor를 제거하는 함수

def reduce_feature(X, column):
    
    # notice
    print("remove top feature {}".format(column))
    
    # make new df
    X = X.drop([column], axis=1)

    # make new model and summary
    model = sm.OLS(y, X).fit()
    # print(model.summary())
    print(model.pvalues[model.pvalues <= 0.01])
    
    # check VIF
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif["feature"] = X.columns
    vif.sort_values(by="VIF Factor", ascending=False, inplace=True)
    print(vif)
    
    top_feature = vif['feature'][0]
    print("\n")
    
    return X, top_feature
```

```python
top_feature = "인버터 1_S상 전류"
for _ in range(5):
    X, top_feature = reduce_feature(X, top_feature)
```

### 모델링 (RandomForest)

```python
data_revise = data_solar.copy()
data_revise.drop(['공통_일사량', '공통_총 전력(AC)'], axis=1, inplace=True)

# x, y 데이터셋 나누기
X = data_revise.drop(['오차'], axis=1)
y = data_revise.iloc[:, -1:]

# 스케일링 (위에서 임포트한 MinMax Scaler 사용)
X_data = scaler.fit_transform(X)
X = pd.DataFrame(data=X_data, columns=X.columns, index=X.index)

y_data = scaler.fit_transform(y)
y = pd.DataFrame(data=y_data, columns=y.columns, index=y.index)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X, y)
```

```python
feature_importance = pd.DataFrame(columns=['feature', 'importance'])
feature_importance['feature'] = X.columns
feature_importance['importance'] = rf.feature_importances_
feature_importance.sort_values(by='importance', ascending=False)
```
